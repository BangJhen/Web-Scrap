```python
import pandas as pd
import numpy as np
import math
pd.set_option('future.no_silent_downcasting', True)
```

```python
df = pd.read_csv("./Provinsi_Jakarta5.csv")
df.head()
```

```python
df.to_json("Provinsi_Jakarta.json", orient="records")
```

```python
df["kelipatan"] = df["Harga"].apply(lambda x : x.split(" ")[-1])
df["Harga"] = df["Harga"].apply(lambda x : x.split(" ")[1].replace(",", ".")).astype(np.float64) 
```

```python
df["kelipatan"] = df["kelipatan"].replace({"Miliar" : math.pow(10, 9), "Juta" : math.pow(10,6)})
```

```python
df["Harga"] = df["Harga"] * df["kelipatan"]
```
